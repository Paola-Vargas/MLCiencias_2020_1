```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
```

```python
items=['azucar','cafe','galletas','huevos','leche']
data= [['azucar','cafe','huevos'],['azucar','huevos','leche'],
       ['azucar','galletas','huevos'],['azucar','huevos'],['azucar','galletas'],
       ['azucar','cafe','galletas'],['cafe','huevos'],
       ['azucar','cafe','galletas','huevos'],['cafe','galletas'],
       ['azucar','cafe','leche']]
```

```python
t = TransactionEncoder()
t_e = t.fit(data).transform(data)
df = pd.DataFrame(t_e, columns=t.columns_)
df
```

```python
apriori(df, min_support=0.2)
```

```python
apriori(df, min_support=0.2, use_colnames=True)
```

```python
items = apriori(df, min_support=0.2, use_colnames=True)
items['itemtamanio'] = items['itemsets'].apply(lambda x: len(x))
items
```

```python
items[(items['itemtamanio'] == 2) &(items['support'] >= 0.4) ]
```

```python
items[items['itemsets'] == {'galletas', 'cafe'} ]
```
